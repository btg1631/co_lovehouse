# 분석 가설 1  : 코리빙사이트와 다방 사이트간의 방 형식의 월세, 보증금 분포가 서로 다를것이다
- 평수 1평당 평균 deposit이랑 rentFee를 구해서 비교해 다방이랑 따로 비교를 분산으로 ㄱㄱ -> roomType, gender 별로 비교
- 평균값 차이 그래프
- 그냥 카운트해서 그냥 범주형 그래프프프프

```python
DW_deposit = df_DGDRWOOJOO['deposit'].sum()
DW_rentFee = df_DGDRWOOJOO['rentFee'].sum()
DW_py = df_DGDRWOOJOO['py'].sum()
```

```python
DABANG_deposit = df_DABANG['deposit'].sum()
DABANG_rentFee = df_DABANG['rentFee'].sum()
DABANG_py = df_DABANG['py'].sum()
```

```python
COLIVING_deposit = df_COLIVING['deposit'].sum()
COLIVING_rentFee = df_COLIVING['rentFee'].sum()
COLIVING_py = df_COLIVING['py'].sum()
```

## 1평당 평균
- DGDRWOOJOO : deposit - 367080  /  rentFee - 43324
- DABANG     : deposit - 594373  /  rentFee - 17005
- COLIVING   : deposit - 293792  /  rentFee - 37588

```python
DW_deposit_result = DW_deposit/DW_py
DW_rentFee_result = DW_rentFee/DW_py
print(DW_deposit_result, DW_rentFee_result)
```

```python
DABANG_deposit_result = DABANG_deposit/DABANG_py
DABANG_rentFee_result = DABANG_rentFee/DABANG_py
print(DABANG_deposit_result, DABANG_rentFee_result)
```

```python
COLIVING_deposit_result = COLIVING_deposit/COLIVING_py
COLIVING_rentFee_result = COLIVING_rentFee/COLIVING_py
print(COLIVING_deposit_result, COLIVING_rentFee_result)
```

```python
df_DGDRWOOJOO['deposit_mean'] = df_DGDRWOOJOO['py']*DW_deposit_result
df_DGDRWOOJOO['rentFee_mean'] = df_DGDRWOOJOO['py']*DW_rentFee_result
df_DGDRWOOJOO['deposit_diff'] = df_DGDRWOOJOO['deposit_mean']-df_DGDRWOOJOO['deposit']
df_DGDRWOOJOO['rentFee_diff'] = df_DGDRWOOJOO['rentFee_mean']-df_DGDRWOOJOO['rentFee']
df_DGDRWOOJOO[:3]
```

```python
df_DABANG['deposit_mean'] = df_DABANG['py']*DABANG_deposit_result
df_DABANG['rentFee_mean'] = df_DABANG['py']*DABANG_rentFee_result
df_DABANG['deposit_diff'] = df_DABANG['deposit_mean']-df_DABANG['deposit']
df_DABANG['rentFee_diff'] = df_DABANG['rentFee_mean']-df_DABANG['rentFee']
df_DABANG[:2]
```

```python
df_COLIVING['deposit_mean'] = df_COLIVING['py']*COLIVING_deposit_result
df_COLIVING['rentFee_mean'] = df_COLIVING['py']*COLIVING_rentFee_result
df_COLIVING['deposit_diff'] = df_COLIVING['deposit_mean']-df_COLIVING['deposit']
df_COLIVING['rentFee_diff'] = df_COLIVING['rentFee_mean']-df_COLIVING['rentFee']
df_COLIVING[:3]
```

## roomType 비교
- 0(1인실)/1(그 외)

```python
df_COLIVING['roomType'] = df_COLIVING['roomType'].replace('1인실', '0')
df_COLIVING['roomType'] = df_COLIVING['roomType'].replace('그 외', '1')
```

```python
# deposit/rentFee	평균, 분산, 중앙값, min, max
mean_DGDRWOOJOO = df_DGDRWOOJOO.groupby('roomType')[['deposit_diff', 'rentFee_diff']].agg(
    deposit_mean=('deposit_diff', 'mean'),
    deposit_var=('deposit_diff', 'var'),
    deposit_median=('deposit_diff', 'median'),
    deposit_min=('deposit_diff', 'min'),
    deposit_max=('deposit_diff', 'max'),
    rentFee_mean=('rentFee_diff', 'mean'),
    rentFee_var=('rentFee_diff', 'var'),
    rentFee_median=('rentFee_diff', 'median'),
    rentFee_min=('rentFee_diff', 'min'),
    rentFee_max=('rentFee_diff', 'max')
)
mean_COLIVING = df_COLIVING.groupby('roomType')[['deposit_diff', 'rentFee_diff']].agg(
    deposit_mean=('deposit_diff', 'mean'),
    deposit_var=('deposit_diff', 'var'),
    deposit_median=('deposit_diff', 'median'),
    deposit_min=('deposit_diff', 'min'),
    deposit_max=('deposit_diff', 'max'),
    rentFee_mean=('rentFee_diff', 'mean'),
    rentFee_var=('rentFee_diff', 'var'),
    rentFee_median=('rentFee_diff', 'median'),
    rentFee_min=('rentFee_diff', 'min'),
    rentFee_max=('rentFee_diff', 'max')
)
```

```python
mean_DGDRWOOJOO['name'] = '코리빙-일반사업자'
mean_COLIVING['name'] = '코리빙-브랜드'
```

```python
combined_data_mean = pd.concat([mean_DGDRWOOJOO, mean_COLIVING])
combined_data_mean = combined_data_mean.reset_index()
combined_data_mean['roomType'] = combined_data_mean['roomType'].astype(int)
combined_data_mean
```
