## 데이터 전처리
- py : ㎡ 삭제
- deposit, rentFee : 콤마, 원 삭제
- roomOption : 아이콘, 공용 가전 : , \n, /, ▶ 삭제
- 붙어있는 단어들 -> Okt로 어떻게 안되겟나..

```python
import pandas as pd
```

```python
def remove(series, char):
    return series.replace(char, '')

def remove_space(series, char):
    return series.replace(char, ' ')

def remove_multiple(series, chars):
    for char in chars:
        series = series.str.replace(char, '')
    return series

def changeint(series):
    series['gender'] = series['gender'].replace('공용', '0')
    series['gender'] = series['gender'].replace('남', '1')
    series['gender'] = series['gender'].replace('여', '2')
    series['roomType'] = series['roomType'].replace('1인실', '0')
    series['roomType'] = series['roomType'].replace('그 외', '1')
    series['brandType'] = series['brandType'].replace('일반사업자', '0')
    series['brandType'] = series['brandType'].replace('맹그로브', '1')
    series['brandType'] = series['brandType'].replace('디어스', '2')
    series['brandType'] = series['brandType'].replace('커먼타운', '3')
    series['brandType'] = series['brandType'].replace('하품', '4')
    return series
```

## 동거동락

```python
df_DGDR = pd.read_csv('project_coliving.gatheringdatas1.csv')
```

```python
df_DGDR['py'] = df_DGDR['py'].apply(lambda x: remove(x, '㎡'))
df_DGDR['gender'] = df_DGDR['gender'].apply(lambda x: remove(x, '성'))
df_DGDR['deposit'] = remove_multiple(df_DGDR['deposit'], ['원', ','])
df_DGDR['rentFee'] = remove_multiple(df_DGDR['rentFee'], ['원', ','])
invalid_rows = df_DGDR[df_DGDR['rentFee'] == '-'].index
df_DGDR = df_DGDR.drop(invalid_rows)
```

```python
df_DGDR['roomOption'] = remove_multiple(df_DGDR['roomOption'], ['(건조기 겸용)', '(1000원60분)', '(방 마다 유무차이 있음)', '집기류 :', '세탁실 :', '기타 : ', '기본옵션 :', '개인방 가전:', '공용 가전 :', '부엌 가전 :', '공용 소모품 :', '공용 가구 :', '가전 :', '공용', '가구 :', '가구:', '또는', '거실', '▶', '& ', '/', ',', '.', r'\n'])
df_DGDR['region'] = remove_multiple(df_DGDR['region'], [',', '/', '에서', '(2호선)', '(2호선', '(4호선)', '(7호선)', '(경의중앙선)'])
```

```python
df_DGDR = changeint(df_DGDR)
```

```python
df_DGDR['gender'] = df_DGDR['gender'].astype(int)
df_DGDR['roomType'] = df_DGDR['roomType'].astype(int)
df_DGDR['py'] = df_DGDR['py'].astype(float)
df_DGDR['deposit'] = df_DGDR['deposit'].astype(int)
df_DGDR['rentFee'] = df_DGDR['rentFee'].astype(int)
df_DGDR['brandType'] = df_DGDR['brandType'].astype(int)
```

```python
df_DABANG = pd.read_csv('project_coliving.yous_dabang.csv')
```

```python
df_DABANG['py'] = df_DABANG['py'].apply(lambda x: remove(x, '㎡'))
df_DABANG['deposit'] = df_DABANG['deposit'].str.replace('만원', '0000')
df_DABANG['rentFee'] = df_DABANG['rentFee'].str.replace('만원', '0000')
df_DABANG['deposit'] = df_DABANG['deposit'].apply(lambda x: remove(x, '억'))
df_DABANG['roomOption'] = df_DABANG['roomOption'].apply(lambda x: remove_space(x, r'\n'))
```

```python
df_DABANG = changeint(df_DABANG)
```

```python
df_DABANG['gender'] = df_DABANG['gender'].astype(int)
df_DABANG['roomType'] = df_DABANG['roomType'].astype(int)
df_DABANG['py'] = df_DABANG['py'].astype(float)
df_DABANG['deposit'] = df_DABANG['deposit'].astype(int)
df_DABANG['rentFee'] = df_DABANG['rentFee'].astype(int)
df_DABANG['brandType'] = df_DABANG['brandType'].astype(int)
```

```python
df_DABANG.info()
```

```python
# region -> '동, 읍, 면, 가'만 남기기
import re
def extract_region(region):
    match = re.search(r'\b\w*동\b|\b\w*읍\b|\b\w*면\b|\b\w*가\b', region)
    if match:
        return match.group()
    return None
```

```python
df_DABANG['region'] = df_DABANG['region'].apply(extract_region)
```
