```python
import pandas as pd
```

```python
# get_data = pd.read_csv('./data/DGDR.csv')
get_data = pd.read_csv('./data/WOOJOO.csv')
get_data = get_data.dropna()
```

```python
get_data.query('region.isna()')
```

```python
station_data = pd.read_csv('./data/station_location.csv')
station_data
```

```python
import re
```

```python
def del_gwalho(get_list):
    station_except = []
    for i in get_list:
        i = re.split(r'[()-]', i)[0]
        station_except.append(i)
    return station_except
```

```python
station_data['del_staion'] = del_gwalho(station_data['역명'])
station_data
```

```python
station_data['지번주소'] = station_data['지번주소'].fillna('')
```

```python
def split_data(get_list):
    get_dong = []
    for i in get_list:
        juso = list(i.split())
        for j in juso:
            if j[-1] == '동':
                j = re.sub(r'\d', '', j)
                get_dong.append(j)
                break
            elif j[-1] == '가':
                get_dong.append(j)
                break
        else:
            get_dong.append('')
    return get_dong              
```

```python
station_data['동추출'] = split_data(station_data['지번주소'])
station_data
```

```python
get_data['region'] = del_gwalho(get_data['region'])
```

```python
get_data['station'] = get_data['region'].apply(lambda x: re.sub(r'역.*$', '', x))
```

```python
get_data['region'].unique()
```

```python
def dong_same(get_list,station_list,dong_list):
    same_dong_list = []
    for i in get_list:
        counter = 0
        for j in station_list:
            if i in j:
                same_dong_list.append(dong_list.loc[counter])
                break
            counter += 1
        else:
            if i == '서울대벤처타운':
                same_dong_list.append('신림동')
            elif i == '서울대벤쳐타운':
                same_dong_list.append('신림동')
            elif i == '노량진':
                same_dong_list.append('노량진동')
            elif i == '판교':
                same_dong_list.append('백현동')
            else:
                same_dong_list.append('')
    return same_dong_list
```

```python
get_data['station_region'] = dong_same(get_data['station'],station_data['del_staion'],station_data['동추출'])
```

```python
get_data['station_region'] = del_gwalho(get_data['station_region'])
```

```python
get_data = get_data[get_data['station_region'] != '']
```
