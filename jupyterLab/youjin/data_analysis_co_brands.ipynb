## 데이터 분리 : train과  test set 분리
 - target과 labels(features) 분리
 - train과 test set 분리

```python
target = df_DGDRWOOJOO_merged['py']
features = df_DGDRWOOJOO_merged[['deposit', 'rentFee']]
# 레코드 개수 확인
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.3, random_state=2)  # random_state - 무작위 데이터 선택하는 연산
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

# 모델

## 목표 변수에 따른 모델 선택 - 연속형
 - 목표변수(Target Variable)
 - 설명변수(Feature Variable)

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(features_train, target_train)
```

# 평가

```python
from sklearn.metrics import r2_score
# test dataset
target_predict_test = model.predict(features_test)
target_test.shape, target_predict_test.shape
```

```python
# train data set에 대한 평가
r2_score(target_test, target_predict_test)
```
